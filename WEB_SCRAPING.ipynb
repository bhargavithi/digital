#TASK
SCRAPING ALL CARS

```python
car_make=[]
car_model=[]
price=[]
fuel_type=[]
transmission_type=[]
available_in_colorvariants=[]
location=[]


for i in result:
  # Car Make
    try:
        car_make.append(i.find('h2').get_text())
    except:
        car_make.append('n/a')

  # Car Model
    try:
        car_model.append(i.find('h3').get_text())
    except:
        car_model.append('n/a')

  # Price
    try:
        price.append(i.find('div',{'data-testid':'car_price'}).get_text())
    except:
        price.append('n/a')

  # Fuel Type, Transmission, and Color Variants
    try:
        prop = i.find('div',{'class':"styles__VariantInfo-sc-a6403e05-12 bkhKsh"})
        prop_list = [p.get_text() for p in prop.find_all('p')]
        fuel_type.append(prop_list[0])
        transmission_type.append(prop_list[1])
        available_in_colorvariants.append(prop_list[2])
    except:
        fuel_type.append('n/a')
        transmission_type.append('n/a')
        available_in_colorvariants.append('n/a')


  # Location
    try:
        location.append(i.find('div',{'class':'styles__CityName-sc-a6403e05-17 fZrHFu'}).get_text())
    except:
        location.append('n/a')

car_df = pd.DataFrame({'car_make': car_make, 'car_model': car_model, 'price': price, 'Fuel Type': fuel_type, 'Transmission': transmission_type, 'Color_variants':available_in_colorvariants, 'Location':location})
display(car_df)
```
